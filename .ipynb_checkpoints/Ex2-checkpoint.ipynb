# So sánh số ca nhiễm và hồi phục giữa các châu lục

```python
# Ở đây ta sử dụng 2 biểu đồ cột đè lên nhau để thấy được sự chênh lệch giữa số ca nhiễm và hồi phục

# Vẽ 2 biểu đồ cột chồng lên nhau
plt.figure(figsize = (10, 5))
sns.barplot(x="Continents", y="Total cases", data=World,color='red') 
sns.barplot(x="Continents", y="Total Recovered", data=World,color='b') 
plt.title('Số ca nhiễm và hồi phục ')


# Vì châu Đại Dương có số ca khá nhỏ, không thể hiện rõ nét trên biểu đồ nên em vẽ thêm 1 cột phụ
plt.figure(figsize = (2, 2))
sns.barplot(x="Continents", y="Total cases", data=World[-1:],color='red') 
sns.barplot(x="Continents", y="Total Recovered", data=World[-1:],color='b') 
plt.title('Số ca nhiễm và hồi phục ')

plt.show()
```

Dựa vào biểu đồ trên thì ta thấy rằng châu Phi hiện đang có số ca hồi phục tích cực nhất khi hầu hết ca nhiễm đều được hồi phục, theo sau đó là châu đại dương, châu Á và Nam Mỹ, cuối cùng là châu Âu và Bắc Mỹ khi số ca chưa được chữa khỏi vẫn còn khá lớn.


Ngoài ra, khi nhìn vào biểu đồ trên thì ta thấy rằng châu Á đứng đầu thế giới về số ca nhiễm, như vậy việc chống dịch của châu Á là tệ nhất? Hay đâu mới là châu lục có tỷ lệ nhiễm covid đáng báo động?

# So sánh tỉ lệ số ca nhiễm, số ca tử vong, số ca test covid và tỉ lệ dân số giữa các châu lục

```python
# Ở đây ta so sánh các châu lục với nhau nên để dễ nhìn em sử dụng biểu đồ pie
label = World['Continents']
colors = sns.color_palette('pastel')[0:6]

# Vẽ biểu đồ pie thể hiện tỉ lệ số ca nhiễm covid trên 1 triệu dân
plt.figure(dpi=100)
pie_tot_cases = World['Tot Cases/ 1M pop']
plt.pie(pie_tot_cases,labels=label, colors = colors, autopct='%.0f%%')
plt.title('Tỉ lệ số ca nhiễm covid trên 1 triệu dân')

# Vẽ biểu đồ pie thể hiện tỉ lệ số ca tử vong do covid trên 1 triệu dân
plt.figure(dpi=100)
pie_tot_deaths = World['Deaths/ 1M pop']
plt.pie(pie_tot_deaths,labels=label, colors = colors, autopct='%.0f%%')
plt.title('Tỉ lệ số ca tử vong do covid trên 1 triệu dân')

# Vẽ biểu đồ pie thể hiện tỉ lệ số ca test covid trên 1 triệu dân
plt.figure(dpi=100)
pie_tot_tests = World['Tests/ 1M pop']
plt.pie(pie_tot_tests,labels=label, colors = colors, autopct='%.0f%%')
plt.title('Tỉ lệ số ca test covid trên 1 triệu dân')

# Vẽ biểu đồ pie thể hiện tỉ lệ dân số giữa các châu lục
plt.figure(dpi=100)
pie_population = World['Population']
plt.pie(pie_population,labels=label, colors = colors, autopct='%.0f%%')
plt.title('Tỉ lệ dân số giữa các châu lục')


plt.show()
```

  Như vậy, ta thấy châu Âu là châu có tỉ lệ số ca nhiễm covid và tỉ lệ số ca tử vong vì covid trên 1 triệu dân là lớn nhất(33%,31%), theo sau là:
- Châu Đại Dương (31%,29%)
- Bắc Mỹ (20%,18%)
- Nam Mỹ (17%,10%)
- Châu Phi (3%,3%)
- Châu Á (2%,1%)

 Ta thấy châu Á có tỉ lệ số ca nhiễm và số ca tử vong thấp nhất, dù cho châu Á có số lượng người đông nhất thế giới(59%), tiếp theo là châu Phi(18%), châu Âu(10%), Bắc Mỹ(8%), Nam Mỹ(6%) và châu Đại Dương(1%)
 
 Hai châu đông dân nhất thì lại có tỉ lệ số ca nhiễm và số ca tử vong thấp nhất, có thể là vì diện tích đất liền rộng lớn của cả 2 khiên cho dịch bệnh lây lan khó khăn, thế nên châu Âu và châu Đại Dương với diện tích nhỏ hơn mà lại có tỉ lệ cao nhất, cũng có thể là do tốc độ phản ứng chậm chạp trước đại dịch nên mới để dịch bùng phát.
 
 Trước tình hình như vậy, người dân châu Âu và châu Đại Dương đã có những biện pháp gì? 
 Tỉ lệ số ca test covid của châu Âu chiếm áp đảo biểu đồ(82%). Nếu ta xem trong database ta sẽ biết được châu Âu có 1316200 ca test covid trên 1 triệu dân, có thể thấy hầu như người dân toàn châu Âu cũng đã được xét nghiệm và có những người được xét nghiệm nhiều hơn 2 lần.
 
 Từ những dữ liệu trên ta kết luận rằng nếu có những biện pháp phản ứng kịp thời, thì dù cho có đông đúc người dân thì vẫn có thể kiểm soát được dịch bệnh lây lan, và nếu có để dịch bùng phát thì không bao giờ là quá trễ để có thể kìm hãm đại dịch

# So sánh tỉ lệ ca nhiễm tăng với số ca chết 


```python
# Em dùng barplot để dễ thể hiện tỷ lệ và độ chênh lệch
plt.figure(figsize = (10, 5))
sns.barplot(x="Continents", y="New cases", data=World,color='r') 
plt.title('Số ca mắc mới ')

plt.figure(figsize = (10, 5))
sns.barplot(x="Continents", y="New deaths", data=World,color='red') 
plt.title('Số ca tử vong mới ')


plt.show()
```

# Thể hiện tình trạng covid mỗi thành phố của mỗi châu lục

```python
# Em dùng biểu đồ boxplot để thể hiện trực quan sự phân bố giữa các thành phố trong châu lục
sns.set_theme(style="whitegrid")

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Total cases", data=All_World)
plt.title('Tổng số ca mắc covid')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="New cases", data=All_World)
plt.title('Số ca mắc mới trong ngày ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Total deaths", data=All_World)
plt.title('Tổng số ca tử vong ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="New deaths", data=All_World)
plt.title('Số ca tử vong mới trong ngày ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Total Recovered", data=All_World)
plt.title('Tổng số ca hồi phục ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="New Recovered", data=All_World)
plt.title('Số ca hồi phục mới trong ngày ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Active Cases", data=All_World)
plt.title('Tổng số ca đang được theo dõi ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Serious, Critical", data=All_World)
plt.title('Tổng số ca nguy kịch')


plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Tot Cases/ 1M pop", data=All_World)
plt.title('Số ca mắc trên 1 triệu dân ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Deaths/ 1M pop", data=All_World)
plt.title('Số ca tử vong trên 1 triệu dân ')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Total Tests", data=All_World)
plt.title('Tổng số ca test')

plt.figure(figsize = (10, 5))
ax = sns.boxplot(x="Continents", y="Tests/ 1M pop", data=All_World)
plt.title('Số ca test trên 1 triệu dân ')

plt.show()
```
